```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('Housing.csv')
df
```

```python
%matplotlib inline
plt.x_label=('area in sqfeet')
plt.ylabel=('price')
plt.plot(df.area,df.price,'g+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict(800)
```
