```python
 import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
inputs=df.drop('Survived',axis='columns')
target=df.Survived
```

```python
dumies=pd.get_dummies(inputs.Sex)
dumies.head()
```

```python
inputs=pd.concat([inputs,dumies],axis='columns')
inputs.head()
```

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python

model.score(X_test,y_test)
```

```python
X_test[0:10]
```

```python
y_test[0:10]
```

```python
model.predict(X_test[0:10])
```

```python
model.predict_proba(X_test[:10])
```

```python
import pandas as pd
d=pd. read_csv('spam.csv')
d.head()
```

```python
d.groupby('Category').describe()
```

```python
d['spam']=d['Category'].apply(lambda x: 1 if x=='spam' else 0)
d.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(d.Message,d.spam)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
X_train_count=v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count=v.transform(X_test)
model.score(X_train_count,y_train)
```

```python
from sklearn.pipeline import Pipeline
clf=Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
