```python
import pandas as pd
df=pd.read_csv("/content/salaries.csv")
df.head()
```

```python
inputs=df.drop('salary_more_then_100k',axis='columns')
```

```python
target=df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['company_n']=le_company.fit_transform(inputs["company"])
inputs['job_n']=le_company.fit_transform(inputs["job"])

inputs['degree_n']=le_company.fit_transform(inputs["degree"])
```

```python
inputs
```

```python
inputs_n= inputs.drop(['company','job','degree'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,1,0]])
```

```python
import pandas as pd
d= pd.read_csv('titanic.csv')
d.head()
```

```python
inp=d.drop('Survived',axis='columns')
```

```python
targets=d['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder

le_Pclass=LabelEncoder()
le_Sex=LabelEncoder()
le_Age=LabelEncoder()
le_Fare=LabelEncoder()
```

```python
inp['Pclass_n']=le_Pclass.fit_transform(inp['Pclass'])
inp['Sex_n']=le_Sex.fit_transform(inp['Sex'])
inp['Age_n']=le_Age.fit_transform(inp['Age'])
inp['Fare_n']=le_Fare.fit_transform(inp['Fare'])
```

```python
inp
```

```python
inp_n=inp.drop([	'PassengerId',	'Pclass'	,'Name',	'Sex'	,'Age',	'SibSp'	,'Parch',	'Ticket',	'Fare',	'Cabin',	'Embarked'],axis='columns')
```

```python
inp_n
```

```python
targets
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inp_n, targets)
```

```python
model.score(inp_n,targets)
```

```python
model.predict([[2,1,28,18 ]])
```
