```python
import pandas as pd
import numpy as np
import sklearn.linear_model
```

```python
df=pd.read_csv('/content/train.csv')
```

```python
df.head()


```

```python
d=df[['Mileage','Engine volume','Price']]
```

```python
d=d.sample(n = 20)
d['Mileage'] = d['Mileage'].str.replace(',', '').str.extract('(\d+.\d+|\d+)').astype(float)
d['Mileage'] = pd.to_numeric(d['Mileage'], errors='coerce')
d['Engine volume'] = d['Engine volume'].apply(lambda x: str(x).replace(',', '') if isinstance(x, str) else x)
d['Engine volume'] = d['Engine volume'].str.extract('(\d+.\d+|\d+)')
d['Engine volume'] = pd.to_numeric(d['Engine volume'], errors='coerce')
d
```
